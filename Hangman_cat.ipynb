```python
# your code here

import random

print("1- Automobile and Household Companies\n2- Countries\n3- Cities")
cat = int(input("Choose your category - 1, 2, or 3: "))

if cat == 1:
    word = random.choice(Auto_household_brands)
elif cat == 2:
    word = random.choice(countries)
elif cat == 3:
    word = random.choice(cities)

word = word.upper()


picked_letter_list =[]
hidden_word = "-"*len(word)
print('\n==============================================')
print(f"\nHidden word is {hidden_word} with {len(word)} character\n")
print('==============================================\n')

mistake_count = 6
while mistake_count > 0 and hidden_word.count("-") != 0:
  index_count = 0
  while True:
  #   usrInput = input("Enter a single character as input: ")
    picked_letter=input("Enter a letter ").capitalize()
    if picked_letter in picked_letter_list:
      print("You have already used that letter")
      continue
    else:
      picked_letter_list.append(picked_letter)
    print(f"Picked characters - {picked_letter_list}")
    print('\n==============================================\n')
    if len(picked_letter) == 1 and picked_letter.isalpha() and picked_letter in picked_letter_list:
          break
  #   print("Please enter a single character to continue\n")
    print(f"\nHidden word is {hidden_word}")
    print("\nPlease only enter one character\n")
  for letter in word:
      if letter == picked_letter:
        # Replace the old letter with the new letter
        hidden_word = hidden_word[:index_count] + picked_letter + hidden_word[index_count+1:]
      index_count += 1
  if picked_letter in word :
      print("Correct!\n")
      print(f"Hidden word is {hidden_word}")
      print(f"{hangman[6 - mistake_count]}")
      print(f"Picked characters - {picked_letter_list}")
      print('\n==============================================\n')
  else:
      mistake_count = mistake_count-1
      print(f"Wrong! Number of mistakes left {mistake_count}\n")
      print(hidden_word)
      print(f"{hangman[6 - mistake_count]}")
      print(f"Picked characters - {picked_letter_list}")
      print('\n==============================================\n')
if mistake_count == 0:
  print(f"\nHANGED! Sorry You Lost. The word was {word}\n")
else:
  print("\nCONGRATULATIONS!!!\n")

```
